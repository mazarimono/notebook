### 必要データの抽出・データの前処理
- すべてのデータを使うわけではないので使うデータを抽出し、そのあと前処理します

```
# 必要なデータを抽出します
col_name = ["年代", "性別", "確定日", "居住都道府県", "X", "Y"]

covid_jp = df[col_name].copy()
```

```
covid_jp
```

#### 各列の要素を確認し、処理方針を決めます

```
# 各列データのユニーク要素を確認します
for col in covid_jp.columns[:-2]:
  print(covid_jp[col].unique())
```

#### 前処理方針
- 年代の列はすべて文字列となっている => 数値計算として扱うため数値型にしておきたい
- 不明は雑に日本の平均年齢の50代にしておく。実際はこの辺はケースバイケース。この場合だと感染者の平均年齢にした方が良いかもしれない
- 性別は、男性・女性・不明の三つに統一。男児、女児は置き換える
- 日付データはdatetime型に置き換える。

### mapメソッド、applyメソッド
- pandasにある
  - mapメソッドはSeriesに対して適用される
  - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html
  - applyメソッドは1列全体、1行全体（引数axisに1を渡す）に対して適用される
  - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html#pandas.DataFrame.apply
- それぞれメソッドの第一引数に関数を渡してやるとそれが実行される
  - 例えばすべての列を1列ずつ合計したい場合は、df.apply(sum)でいける（はず）

#### 雑誌ではreplaceを使ってやったけど、なんかほかに良い処理ないかなぁ・・・
- 再帰関数を使って削ってやるってアイデアもあったんですけど、それで対応しきれない可能性も（今回だと"'\xa00-10'"）
- 男・女・不明は正規表現で「男」があると男性に置き換えるなんて処理もありかなぁ
- とか思っていますが、何か良いアイデアとか普段こうするとかあったら教えてください


```
covid_jp["年代"] = covid_jp["年代"].map(lambda x: x.replace("不明", "50"))
```

```
covid_jp["年代"].unique()
```

```
covid_jp["年代"] = covid_jp["年代"].map(lambda x: x.replace("0-10", "0"))
covid_jp["年代"] = covid_jp["年代"].map(lambda x: x.replace("90以上", "90"))
covid_jp["年代"] = covid_jp["年代"].map(lambda x: x.replace("\xa00-10", "0"))
covid_jp["年代"] = covid_jp["年代"].map(lambda x: x.replace('100歳以上', "100"))
covid_jp["年代"] = covid_jp["年代"].map(lambda x: int(x))
```

```
covid_jp["年代"].unique()
```

```
import re
```

```
re.search("男", "女性")
```

```
def update_sex(x):
  if re.search("男", x):
    return "男性"
  elif re.search("女", x):
    return "女性"
  else:
    return x
```

```
covid_jp["性別"].map(lambda x: update_sex(x)).unique()
```

```
covid_jp["性別"] = covid_jp["性別"].apply(lambda x: update_sex(x))
```

```
covid_jp["性別"].unique()
```

```
covid_jp["確定日"] = pd.to_datetime(covid_jp["確定日"])
```

```
covid_jp.info()
```

```
covid_jp.tail()
```

```
covid_jp["count"] = 1
```

```
# データの保存
covid_jp.to_csv("covid_jp.csv")
```

## Plotly Expressによる可視化
- Plotly Expressとはplotlyというグラフパッケージを簡単に使えるパッケージ
- 関数（df, x="列名1", y="列名2"）のように簡単に可視化ができる
- その際にこのデータで重要になるのがデータの集計

### データの集計
- 今回のデータは個別のデータであるため、都道府県のデータや日別のデータを見る際には集計しないといけない
- そのような際に使えるのがgroupbyメソッド
  - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html


```
test_group = covid_jp.groupby("居住都道府県")
```

```
test_group["count"].sum()

test_group_count = pd.DataFrame(test_group["count"].sum())
```

```
test_group["年代"].mean()
test_group_age = pd.DataFrame(test_group["年代"].mean())
test_group_count.merge(test_group_age, left_index=True, right_index=True)
```

```
count_and_mean_age = test_group_count.merge(test_group_age, left_index=True, right_index=True)
count_and_mean_age = count_and_mean_age.reset_index()
count_and_mean_age
```

```
import plotly.express as px
```

```
count_and_mean_age = count_and_mean_age.sort_values("年代")
px.bar(count_and_mean_age, x="居住都道府県", y="年代")
```

```
# 居住都道府県：鳥取、年代20以下でデータフレームをフィルタする
covid_jp[(covid_jp["居住都道府県"] == "島根県") & (covid_jp["年代"] < 20)]
```

### groupbyでもっと詳細に
- byのところに列名を渡すことにより、条件を付けくわえられる

```
data = covid_jp.groupby(["年代", "性別", "居住都道府県"]).sum()
data = data.reset_index()
data = data.drop(["X", "Y"], axis=1)
data
```

```
data_kyoto = data[data["居住都道府県"] == "京都府"]
px.pie(data_kyoto, names="性別", values="count")
```

```
fig = px.treemap(data, path=["居住都道府県", "性別", "年代"], values="count")
fig.show()
```

### Dashとは
- react, plotly, flaskを組み合わせて作られたウェブフレームワーク
- クラスから作れるコンポーネントを組み合わせてレイアウトを作成します
- コンポーネントは色々提供されていますが、本で用いたのはHTML要素を作るdash_html_componentsと様々なツールを作成するdash_core_componentsです
- それらを組み合わせ、グラフを動的に動かせるのがDashの特徴です。
- グラフを動的に動かすのがコールバックです
- コールバックは何かをきっかけに呼び出され、関数を実行して見た目に変化を生み出します
- https://dash.plotly.com/

```
from jupyter_dash import JupyterDash 
import dash_core_components as dcc 
import dash_html_components as html 
from dash.dependencies import Input, Output
```

```
# 基本的な作り方
app = JupyterDash(__name__) #JupyterDashインスタンスの生成

app.layout = html.H1("みなさんこんばんは！") # レイアウトにコンポーネントを渡す。文字は大体引数children(第1引数)に渡す。（children属性）

app.run_server(mode="inline") # サーバをJupyter上で起動
```
